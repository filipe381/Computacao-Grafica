```python
import numpy as np
import matplotlib.pyplot as plt

def plot_points(original, transformed, title):
    plt.figure(figsize=(6, 6))
    plt.scatter(original[0], original[1], color='blue')
    plt.scatter(transformed[0], transformed[1], color='red')

    plt.title(title)
    plt.xlabel("Eixo X")
    plt.ylabel("Eixo Y")
    plt.legend()
    plt.grid(True)
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

P = np.array([2, 3])
v = np.array([4, -2])
P_prime = P + v

plot_points(P, P_prime,"Exercício: 1")

```

```python
def plot_shapes(original, transformed, title):
    plt.figure(figsize=(8, 8))
    original_closed = np.vstack([original, original[0]])
    transformed_closed = np.vstack([transformed, transformed[0]])

    plt.plot(original_closed[:, 0], original_closed[:, 1], 'o-', label='Original', color='blue')
    plt.plot(transformed_closed[:, 0], transformed_closed[:, 1], 'o--', label='Transformado', color='red')

    plt.title(title)
    plt.xlabel("Eixo X")
    plt.ylabel("Eixo Y")
    plt.legend()
    plt.grid(True)
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

triangle = np.array([
    [1, 1],
    [3, 1],
    [2, 4]
])

scale_factor = 2
scaled_triangle = triangle * scale_factor

plot_shapes(triangle, scaled_triangle, "Exercício 2")

print("Coordenadas originais:")
print(triangle)
print("\nNovas coordenadas após a escala:")
print(scaled_triangle)
```

```python
def plot_points(original, transformed, title):
    plt.figure(figsize=(6, 6))
    plt.scatter(original[0], original[1], color='blue', label='Original P')
    plt.scatter(transformed[0], transformed[1], color='red', label="Transformado P'")
    plt.arrow(original[0], original[1], transformed[0] - original[0], transformed[1] - original[1],
              head_width=0.2, head_length=0.2, fc='gray', ec='gray', linestyle='--')
    plt.title(title)
    plt.xlabel("Eixo X")
    plt.ylabel("Eixo Y")
    plt.legend()
    plt.grid(True)
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

P = np.array([1, 0])
angle_degrees = 90
angle_radians = np.deg2rad(angle_degrees)

c, s = np.cos(angle_radians), np.sin(angle_radians)
rotation_matrix = np.array([
    [c, -s],
    [s, c]
])

P_prime = rotation_matrix.dot(P)

plot_points(P, P_prime, "Exercício 4")

print(f"O ponto original P era: {P}")
print(f"O novo ponto P' é: {np.round(P_prime, 2)}")
```

```python
def plot_points(original, transformed, title):
    plt.figure(figsize=(6, 6))
    plt.scatter(original[0], original[1], color='blue', label='Original P')
    plt.scatter(transformed[0], transformed[1], color='red', label="Transformado P'")
    plt.arrow(original[0], original[1], transformed[0] - original[0], transformed[1] - original[1],
              head_width=0.2, head_length=0.2, fc='gray', ec='gray', linestyle='--')
    plt.title(title)
    plt.xlabel("Eixo X")
    plt.ylabel("Eixo Y")
    plt.legend()
    plt.grid(True)
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

P = np.array([3, 2])
v = np.array([1, -1])
P1 = P + v

angle_rad = np.deg2rad(90)
c, s = np.cos(angle_rad), np.sin(angle_rad)
rotation_matrix = np.array([[c, -s], [s, c]])
P2 = rotation_matrix.dot(P1)

scale_factor = 2
P_prime = P2 * scale_factor

plot_points(P, P_prime, "Exercício 9: Composição de Transformações")

print(f"Ponto Inicial P: {P}")
print(f"1. Após Translação P1: {P1}")
print(f"2. Após Rotação P2: {np.round(P2, 2)}")
print(f"3. Ponto Final P': {np.round(P_prime, 2)}")
```

```python
def plot_shapes(original, transformed, title):
    plt.figure(figsize=(8, 8))
    original_closed = np.vstack([original, original[0]])
    transformed_closed = np.vstack([transformed, transformed[0]])
    plt.plot(original_closed[:, 0], original_closed[:, 1], 'o-', label='Original', color='blue')
    plt.plot(transformed_closed[:, 0], transformed_closed[:, 1], 'o--', label='Transformado', color='red')
    plt.title(title)
    plt.xlabel("Eixo X")
    plt.ylabel("Eixo Y")
    plt.legend()
    plt.grid(True)
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

rectangle = np.array([
    [1, 1],
    [5, 1],
    [5, 3],
    [1, 3]
])

v = np.array([-2, 3])
translated_rect = rectangle + v

scale_factors = np.array([1.5, 0.5])
scaled_rect = translated_rect * scale_factors

reflection_matrix_y = np.array([[-1, 0], [0, 1]])
final_rect = reflection_matrix_y.dot(scaled_rect.T).T

plot_shapes(rectangle, final_rect, "Exercício 10: Combinação de Transformações")

print("Coordenadas originais:")
print(rectangle)
print("\nCoordenadas finais após todas as transformações:")
print(final_rect)
```
